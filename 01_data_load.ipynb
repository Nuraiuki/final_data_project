### 4.5 — `cities` и `countries`

```python
print('=== cities ===')
display(cities.head())
print('Пропуски:', cities.isnull().sum().sum())

print('\n=== countries ===')
display(countries.head())
print('Пропуски:', countries.isnull().sum().sum())
```

```python
def missing_summary(df, name):
    """Возвращает сводку пропусков в таблице."""
    missing = df.isnull().sum()
    pct     = (missing / len(df) * 100).round(2)
    result  = pd.DataFrame({'Таблица': name, 'Пропуски': missing, '%': pct})
    return result[result['Пропуски'] > 0]

tables = {
    'sales': sales,
    'customers': customers,
    'products': products,
    'categories': categories,
    'cities': cities,
    'countries': countries
}

summary_parts = [missing_summary(df, name) for name, df in tables.items()]
missing_all   = pd.concat(summary_parts)

if missing_all.empty:
    print('✅ Пропусков не обнаружено ни в одной таблице!')
else:
    display(missing_all)
```

## 6. Очистка таблицы `sales`

> Удаляем дубликаты, строки с критически важными пропусками, некорректные даты и нулевые/отрицательные значения количества.

```python
print(f'Строк до очистки: {len(sales):,}')

sales = sales.drop_duplicates()
print(f'После drop_duplicates: {len(sales):,}')

sales = sales.dropna(subset=['CustomerID', 'ProductID', 'SalesDate'])
print(f'После dropna (ключевые поля): {len(sales):,}')

sales['SalesDate'] = pd.to_datetime(sales['SalesDate'], errors='coerce')
sales = sales.dropna(subset=['SalesDate'])
print(f'После парсинга дат: {len(sales):,}')

sales = sales[sales['Quantity'] > 0]
print(f'После фильтра Quantity > 0: {len(sales):,}')

print('\nОчистка завершена')
```

---
## 7. Создание признака выручки (Feature Engineering)

> Я пересчитала выручку самостоятельно, чтобы не зависеть от готового столбца `TotalPrice` и показать процесс feature engineering.
>
> Цена берётся из таблицы `products`, так как в `sales` её нет. Затем применяется скидка по формуле:
> **Revenue = Quantity × Price × (1 − Discount)**

```python
# Присоединяем цену из products
sales = sales.merge(
    products[['ProductID', 'Price']],
    on='ProductID',
    how='left'
)

price_missing = sales['Price'].isnull().sum()
print(f'Строк без цены после merge: {price_missing}')

sales['Revenue'] = sales['Quantity'] * sales['Price'] * (1 - sales['Discount'].fillna(0))

print('\n Столбец Revenue добавлен!')
display(sales[['SalesID', 'CustomerID', 'ProductID', 'Quantity', 'Price', 'Discount', 'Revenue']].head(10))
```

## 8. Итоговая проверка и сохранение очищенных данных

```python
print('📊 Итоговая статистика очищенной таблицы sales:')
display(sales[['Quantity', 'Price', 'Discount', 'Revenue']].describe())

print(f'\nДиапазон дат: {sales["SalesDate"].min().date()} → {sales["SalesDate"].max().date()}')
print(f'Уникальных клиентов: {sales["CustomerID"].nunique():,}')
print(f'Уникальных товаров:  {sales["ProductID"].nunique():,}')
print(f'Всего транзакций:    {len(sales):,}')
print(f'Общая выручка:       {sales["Revenue"].sum():,.2f}')
```

```python
sales.to_csv('sales_clean.csv', index=False)
print('✅ sales_clean.csv сохранён')

customers.to_csv('customers_clean.csv', index=False)
products.to_csv('products_clean.csv', index=False)
print('✅ customers_clean.csv и products_clean.csv сохранены')
```
